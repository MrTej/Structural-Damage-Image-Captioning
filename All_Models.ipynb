## Literature Review
The generation of textual descriptions for visual contents or images commonly referred to as image captioning has been a subject of interest in the last few years owing to its possible usability especially in areas like assistive, indexing, and post-disaster reconnaissance. The main objective is to mimic free-form descriptions with the help of the artificial models to process the content of images and relay it in words. In principle, there are mainly three forms of IC: template-based, retrieval-based, and neural-network based.
Template-based models use pre-defined structures of the sentence along with the possible visual prompts like the objects, their features (activity or positioning). These models normally work with predefined rules or templates and there are an advantage of having total control on them. Nonetheless, they have limited variability in the length or context of captions and many articles don’t have maximal descriptive richness (Wang, C., Yang, H. and Meinel, C. (2018)).

Retrieval-based models produce captions for a specific image by searching for images from a set of images that have already been captioned and then translating or aggregating the captions in a final image. These methods are better at generating state-wise, finer-grained, and fluency, better than the templates because they use human-written references. Despite this, rediscovery has two disadvantages inherent to the retrieval-based methods: The first comes with the type and quality of the corpus on which the index training is based, and the second is related to the computational cost, which increases dramatically for large corpora (Xu, K., Wang, H. and Tang, P. (2017)).

The most used method in image captioning now is the use of neural networks due to the developments in both computer vision and natural language processing. These models are often end-to-end ones (Omri, M. et al. (2022)) which employ Convolutional Neural Network as the first stage of an encoder–decoder models (Onita, D., Birlutiu, A. and Dinu, L.P. (2020)) for feature extraction of an image and Recurrent Neural Network Oluwasammi, A. et al. (2021) as a decoder for generating the textual description of the image. it is possible to construct other related syntactically variant captions that would be semantically relevant to a given image. To improve this architecture, more sophisticated framework such as the Recurrent Fusion Network (RFNet) has been presented that uses several CNNs as the encoder which brings in features that are richer semantically for the fusion module to aggregate in its generation of the caption.

Since also the tasks like machine translation and conditional image generation have provoked the use of CNNs as decoders (Oluwasammi, A. et al. (2021) and Jiang, W. et al. (2018)), the current study also adopted this trend. As such, the capability called the attention mechanism has become particularly important. Attention also assists in boosting the two criterion, accuracy and context since models are enabled to focus on the right parts of the image during the captioning task. The image captioning attention mechanisms can be categorized into several type: channel attention, spatial attention, temporal attention and branch attention (Guo, M.-H. et al. (2022)).

For example, Mozaffari, M.H. and Lee, W.-S. (2020) proposed the IrisNet model, which includes a RetinConv block that is designed for the peripheral vision-based filtering of the image to collectively focus on the most important feature areas. Other forms of attention that have been found work for the captioning task include soft and hard attention (Xu, K. et al. (2015)), adaptive attention (Lu, J. et al. (2017)), and top-down and bottom-up attention (Anderson, P. et al. (2018)) presented a spatial and channel-wise attention CNN that augments the attention at various features map levels to include the right focus on context required in descriptive captioning.

Altogether these innovations have significantly enhanced the readability and accuracy as well as the pertinence and ensuing descriptive of automatically created image captions. However, there are issues that still exist and persist, especially when the model is applied in specific domains for instance, structural-damage image captioning; requires the model to comprehend technical jargon and slight features that maximally influence the visuals.

### Reason for Selecting FCN, MobileNet, U-Net, and SegNet for Damage Classification

In the implementation of structural-damage image captioning, selecting appropriate deep learning models is critical for accurately identifying, localizing, and describing the extent of damage in images. The chosen models—**Fully Convolutional Network (FCN)**, **MobileNet**, **U-Net**, and **SegNet**—were selected from among various alternatives based on their architectural strengths, computational efficiency, and proven effectiveness in image segmentation and classification tasks relevant to structural damage analysis. Below are the reasons for selecting each model:

---

#### **1. Fully Convolutional Network (FCN)**

**Reason for selection:**
FCNs are particularly well-suited for **dense prediction tasks** such as semantic segmentation, where pixel-level classification is required. Since structural-damage image captioning demands precise localization of damaged regions to inform descriptive output, FCNs offer a strong foundation. Their end-to-end architecture replaces fully connected layers with convolutional layers, enabling the model to process images of arbitrary sizes while maintaining spatial hierarchies. FCNs are also computationally efficient and have been widely used in medical and structural image segmentation tasks, demonstrating high accuracy in boundary detection.

---

#### **2. MobileNet**

**Reason for selection:**
MobileNet was selected for its **lightweight and efficient architecture**, making it ideal for real-time damage assessment and deployment on **resource-constrained environments**, such as mobile devices or drones. Built on **depthwise separable convolutions**, MobileNet significantly reduces the number of parameters and computational cost without drastically compromising performance. This makes it an excellent backbone for damage classification and captioning pipelines that require fast inference with limited hardware. Additionally, MobileNet's flexibility allows for integration with attention or captioning modules in the later stages of the model pipeline.

---

#### **3. U-Net**

**Reason for selection:**
U-Net is a widely adopted architecture in the field of **biomedical and structural image segmentation**, known for its **encoder-decoder structure with skip connections**. These skip connections help retain fine-grained spatial details lost during downsampling, which is essential for capturing small cracks or subtle structural anomalies in damaged buildings or infrastructure. Its strong performance on small datasets and adaptability to multi-class segmentation make it ideal for structural-damage localization prior to generating textual descriptions. The final classification or captioning layers can be efficiently built on top of U-Net’s rich feature maps.

---

#### **4. SegNet**

**Reason for selection:**
SegNet was chosen for its **robustness in segmenting urban and structural environments**, especially where background noise or varying lighting conditions are present. It uses a similar encoder-decoder approach to U-Net but uniquely **stores pooling indices** during encoding and reuses them in decoding, leading to more accurate and computationally efficient upsampling. This allows SegNet to excel at **class-preserving segmentation**, which is particularly beneficial when distinguishing between different types of structural damage. Its ability to produce clean segmentation maps makes it a strong candidate for feature extraction and region proposal in damage captioning systems.

---


The selection of FCN, MobileNet, U-Net, and SegNet is a strategic balance between **accuracy**, **efficiency**, and **generalizability**. FCN and U-Net provide high-resolution spatial understanding, SegNet ensures efficient and accurate reconstruction of segmentation maps, and MobileNet offers portability and speed for real-time applications. Together, these models provide a comprehensive foundation for building an effective structural-damage image captioning system that meets the needs of both performance and deployment feasibility.

### Image Augmentation and Dataset Balancing for Structural Damage Classification

The below code focuses on image preprocessing and augmentation to balance the dataset across three classes: Damaged, Mild Damage, and No Damage. Class imbalance stands as a usual challenge in real-world datasets where certain image categories outnumber others. Artificial image augmentation serves as a solution within the code to boost rare class samples which promotes model learning and reduces bias.

ImageDataGenerator from Keras defines a data augmentation approach to randomly perform rotation and shifting and zooming and horizontal flipping on existing images. The creation of new slightly transformed versions of original images helps models develop stronger robustness and generalized performance.

The code runs 600 non-augmented image selections from the No Damage category to maintain existing sample levels. The code starts by loading existing images for both the Damaged and Mild Damage categories before adding augmented versions of these pictures so each class contains exactly 500 images total. Having uniform sizes among all three classes becomes essential for creating an unbiased and effective classification model.

The visualization relies on two custom programming functions as a tool. The first function shows random examples from new image data along with both the Damaged and Mild Damage categories to demonstrate the impact of augmentation. The second visualization screens original No Damage class images alongside both augmented and real samples for comparison. Realistic and training-responsive augmented data needs visual confirmation as part of model development.

```python
#image augumentation
# Preprocessing: Augment images to balance the dataset
datagen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
def augment_images(image_paths, num_augmented, target_size=(256, 256)):
    """Augment images to reach the specified total number (original + augmented)."""
    augmented_images = []
    for image in image_paths:
        img = tf.keras.preprocessing.image.load_img(image, target_size=target_size)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        counter = 0
        for batch in datagen.flow(img_array, batch_size=1):
            augmented_images.append(batch[0])
            counter += 1
            if counter + len(image_paths) >= num_augmented:
                break
    return np.array(augmented_images[:num_augmented - len(image_paths)])
```

```python
# Step 1: Randomly select 600 images from No Damage (no augmentation)
no_damage_images = [os.path.join(no_damage_path, img) for img in os.listdir(no_damage_path)]
selected_no_damage = random.sample(no_damage_images, 600)
no_damage_data = [tf.keras.preprocessing.image.img_to_array(tf.keras.preprocessing.image.load_img(img, target_size=(256, 256))) for img in selected_no_damage]
no_damage_data = np.array(no_damage_data)
```

```python
# Step 2: Prepare Damaged class (original + augmented = 500)
damaged_images = [os.path.join(damaged_path, img) for img in os.listdir(damaged_path)]
augmented_damaged = augment_images(damaged_images, num_augmented=500)
damaged_data = np.concatenate([np.array([tf.keras.preprocessing.image.img_to_array(tf.keras.preprocessing.image.load_img(img, target_size=(256, 256))) for img in damaged_images]), augmented_damaged])
```

```python
# Step 3: Prepare Mild Damage class (original + augmented = 500)
mild_damage_images = [os.path.join(mild_damage_path, img) for img in os.listdir(mild_damage_path)]
augmented_mild_damage = augment_images(mild_damage_images, num_augmented=500)
mild_damage_data = np.concatenate([np.array([tf.keras.preprocessing.image.img_to_array(tf.keras.preprocessing.image.load_img(img, target_size=(256, 256))) for img in mild_damage_images]), augmented_mild_damage])
```

### Train and Test Split

The chosen **split ratio** — 70% training, 15% validation, and 15% testing — is a **well-balanced and commonly used strategy** in machine learning for the following reasons:

### 1. **70% for Training:**

* The **training set** is the largest portion (70%) because it is used to **fit the model**—i.e., to learn patterns, features, and relationships in the data.
* More training data helps the model **generalize better** and reduces the risk of overfitting, especially in image classification tasks with deep learning.

### 2. **15% for Validation:**

* The **validation set** is used to **tune hyperparameters** and monitor the model’s performance during training.
* It helps in **early stopping** and **model checkpointing**, ensuring that the model doesn’t overfit the training data.
* A 15% split offers a **sufficiently large yet computationally efficient** sample for validation without sacrificing too much training data.

### 3. **15% for Testing:**

* The **test set** evaluates the **final model’s performance** after training and validation.
* It must be **completely unseen** during training to ensure a **fair and unbiased assessment** of the model’s generalization ability.
* 15% provides a **statistically reliable estimate** of accuracy, precision, recall, and other performance metrics.

### Why Not Other Ratios?

* **80-10-10** might give more training data but less validation/test data, which could reduce confidence in model evaluation.
* **60-20-20** could give more evaluation confidence but less training data, possibly hurting model learning.
* **50-25-25** is usually not ideal unless working with a very large dataset or prioritizing evaluation over learning.

In this case, with a moderately sized dataset (\~1600 images), the **70/15/15 split is optimal**, offering a good trade-off between **learning capability and evaluation robustness**.

### Training a Fully Convolutional Network (FCN) for Structural Damage Classification

This section demonstrates the process of building, training, and evaluating a Fully Convolutional Network (FCN) model using a pretrained ResNet50 backbone to classify structural damage from images into three categories: Damaged, No Damage, and Mild Damage.

First, the model is defined using ResNet50 as the feature extractor (backbone), excluding its top layers to allow customization for the classification task. A series of convolutional and transposed convolutional layers are added for feature refinement and upsampling. However, since this is a classification task and not segmentation, a GlobalAveragePooling2D layer is used before the final dense softmax layer to make predictions at the image level.

The model is compiled using the Adam optimizer and categorical crossentropy loss, appropriate for multi-class classification. A ModelCheckpoint callback is added to save the model that performs best on the validation set, ensuring efficient training and model selection.

The model is then trained on the preprocessed and balanced training data for 20 epochs with a batch size of 16, and its performance is validated on the validation set.

After training, the model's performance is assessed on the test set using several evaluation metrics:
1.   Accuracy
2.   F1 Score
3. Classification Report
4. Confusion Matrix

These metrics offer a clear picture of how well the model generalizes to unseen data.

To aid interpretation:

1. A plot of training/validation accuracy and loss over epochs is provided to analyze the learning process.

2. A bar chart displays the final evaluation metrics (Accuracy and F1 Score).

3. A confusion matrix heatmap shows detailed class-wise prediction performance, helping identify which classes are often confused by the model.

This comprehensive setup ensures that the FCN model is not only trained effectively but also evaluated thoroughly to ensure reliable structural damage classification.

```python
# Define FCN model using a pretrained backbone (e.g., ResNet50)
def fcn_model(input_shape=(256, 256, 3), num_classes=3):
    # Use ResNet50 as the backbone
    base_model = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)

    # Add layers for segmentation
    x = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(base_model.output)
    x = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = tf.keras.layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)
    x = tf.keras.layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same', activation='relu')(x)

    # Change here to GlobalAveragePooling2D for image-level prediction
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

    return tf.keras.models.Model(inputs=base_model.input, outputs=x)
```

```python
# Compile the FCN model
model = fcn_model()
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Add ModelCheckpoint to save the model at the best epoch (based on validation accuracy)
checkpoint = ModelCheckpoint(
    filepath='best_model.h5',        # Filepath to save the model
    monitor='val_accuracy',          # Metric to monitor (change to 'val_loss' if needed)
    save_best_only=True,             # Save only the best model
    mode='max',                      # Mode: 'max' for accuracy, 'min' for loss
    verbose=1                        # Verbosity (1 = progress bar displayed)
)
```

```python
# Train the model with the ModelCheckpoint callback
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=20,
    callbacks=[checkpoint]           # Include the callback
)
```

```python
# Evaluate on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=-1)
y_true = np.argmax(y_test, axis=-1)
```

```python
# Metrics
accuracy = accuracy_score(y_true.flatten(), y_pred_classes.flatten())
f1 = f1_score(y_true.flatten(), y_pred_classes.flatten(), average='weighted')

print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print("Classification Report:")
print(classification_report(y_true.flatten(), y_pred_classes.flatten()))
print("Confusion Matrix:")
cm = confusion_matrix(y_true.flatten(), y_pred_classes.flatten())
print(cm)
```

### Image Augmentation for Dataset Balancing

In the below code, we implement **image augmentation** as part of the preprocessing step to balance the dataset for training. Image augmentation is used to artificially expand the dataset by generating new images through various transformations, such as rotations, shifts, zooms, and flips. This technique is particularly useful when the dataset is imbalanced, ensuring that the model doesn't become biased towards overrepresented classes.

1. **ImageDataGenerator**: The `ImageDataGenerator` from Keras is instantiated with several transformations:

   * **rotation\_range**: Random rotations of up to 25 degrees are applied to images.
   * **width\_shift\_range** and **height\_shift\_range**: The images are randomly shifted along the width and height by 20%.
   * **zoom\_range**: The images are randomly zoomed by 20%.
   * **horizontal\_flip**: Random horizontal flipping of images is applied.

   These transformations help to generate varied images from the original dataset, which improves the model’s robustness.

2. **augment\_images Function**: This function takes a list of image file paths and applies the transformations until the desired number of augmented images is generated. The images are loaded, converted into arrays, and expanded to match the input shape expected by the model. The `datagen.flow()` method generates batches of augmented images, and the function ensures that the total number of images matches the target count.

3. **Handling the Classes**:

   * **No Damage**: Since no augmentation is applied to the **No Damage** class, a random selection of 600 images is chosen from this folder to avoid any class imbalance.
   * **Damaged** and **Mild Damage**: Both of these classes undergo augmentation. For each class, original images are combined with their augmented counterparts to create a total of 500 images per class.

4. **Displaying Augmented Images**: The `plot_augmented_samples()` function visualizes random samples of augmented images from the **Damaged** and **Mild Damage** classes. This helps to verify the effectiveness of the augmentations in increasing the variability and quantity of images in the dataset. For each augmented class, a few randomly chosen images are displayed with the transformations applied.

5. **Non-Augmented Samples**: The `plot_non_augmented_samples()` function is used to visualize random samples from the **No Damage** class, which has not been augmented. This helps in comparing the original images with the augmented ones and ensuring the diversity introduced by augmentation.

By augmenting the dataset, we aim to create a more balanced and robust training set, which is crucial for preventing overfitting and improving the generalization ability of the model. The augmented dataset is expected to better represent real-world variability in the structural damage classes, leading to improved model performance during training.


```python
#Image Augumentation
# Preprocessing: Augment images to balance the dataset
datagen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
def augment_images(image_paths, num_augmented, target_size=(256, 256)):
    """Augment images to reach the specified total number (original + augmented)."""
    augmented_images = []
    for image in image_paths:
        img = tf.keras.preprocessing.image.load_img(image, target_size=target_size)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        counter = 0
        for batch in datagen.flow(img_array, batch_size=1):
            augmented_images.append(batch[0])
            counter += 1
            if counter + len(image_paths) >= num_augmented:
                break
    return np.array(augmented_images[:num_augmented - len(image_paths)])
```

```python
# Step 1: Randomly select 600 images from No Damage (no augmentation)
no_damage_images = [os.path.join(no_damage_path, img) for img in os.listdir(no_damage_path)]
selected_no_damage = random.sample(no_damage_images, 600)
no_damage_data = [tf.keras.preprocessing.image.img_to_array(tf.keras.preprocessing.image.load_img(img, target_size=(256, 256))) for img in selected_no_damage]
no_damage_data = np.array(no_damage_data)
```

### SegNet Model Training and Evaluation with ModelCheckpoint

The below code builds and trains a **SegNet**-like architecture, which is typically used for image segmentation tasks, but in this case, it is adapted for classification. Here's a detailed breakdown:

1. **Model Architecture**:

   * The model consists of multiple **Conv2D** layers followed by **MaxPooling2D** layers to extract hierarchical features from the input images. The first set of convolutional layers uses 64, 128, and 256 filters, respectively, with a **ReLU** activation function and **'same'** padding to maintain the spatial dimensions.
   * The network then applies **Conv2DTranspose** layers (deconvolution layers) to upsample the feature maps, progressively reducing the depth of the feature maps but increasing the spatial resolution.
   * After the upsampling layers, a **GlobalAveragePooling2D** layer is applied to aggregate the feature maps into a single vector, which is then passed to a **Dense** layer with a softmax activation function to produce the final class probabilities (three classes: Damaged, No Damage, Mild Damage).

2. **Model Compilation**:

   * The model is compiled with the **Adam optimizer**, **categorical crossentropy** loss (for multiclass classification), and **accuracy** as the evaluation metric.

3. **ModelCheckpoint Callback**:

   * To ensure the best model is saved during training, a **ModelCheckpoint** callback is used. This callback monitors the validation accuracy (`val_accuracy`) and saves the model with the highest validation accuracy to a file named `best_model3.h5`. The `save_best_only=True` option ensures that only the best-performing model (based on validation accuracy) is saved, helping prevent overfitting.

4. **Model Training**:

   * The model is trained for 20 epochs with a batch size of 16, using the **X\_train** and **y\_train** data for training and **X\_val** and **y\_val** for validation. The training process is monitored by the `ModelCheckpoint` callback to save the best model during the training process.

5. **Plotting Training and Validation Metrics**:

   * After training, the accuracy and loss for both the training and validation datasets are plotted across the epochs to visualize the model's learning process. These plots help identify overfitting or underfitting by comparing the training and validation curves.

6. **Model Evaluation**:

   * The model's performance is evaluated on the **test set** (**X\_test** and **y\_test**). The model's predictions are compared with the true labels to compute **accuracy** and **F1-score**. The **F1-score** is calculated with a weighted average to account for class imbalances.
   * A **classification report** is printed to show precision, recall, and F1-score for each class. Additionally, a **confusion matrix** is generated to visualize the classification performance across the three classes. This matrix helps identify whether the model is misclassifying certain classes.

7. **Confusion Matrix Visualization**:

   * The confusion matrix is visualized using a **heatmap**. This provides a clear view of how the model is performing across all classes (Damaged, No Damage, and Mild Damage), highlighting where misclassifications are occurring.

In summary, this code trains a **SegNet-based** model for structural damage classification, saves the best-performing model during training, and evaluates its performance on the test data using accuracy, F1-score, classification report, and a confusion matrix. It also visualizes the training process and the evaluation metrics to better understand the model's performance.


```python
from tensorflow.keras.callbacks import ModelCheckpoint

def segnet(input_size=(256, 256, 3), num_classes=3):
    inputs = tf.keras.Input(input_size)

    x = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)
    x = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)
    x = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)

    x = tf.keras.layers.Conv2DTranspose(256, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)
    x = tf.keras.layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)
    x = tf.keras.layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)

    gap = tf.keras.layers.GlobalAveragePooling2D()(x)
    outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(gap)
    return tf.keras.models.Model(inputs, outputs)

model3 = segnet()
model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Add ModelCheckpoint
checkpoint3 = ModelCheckpoint(
    filepath='best_model3.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

history3 = model3.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=20,
    callbacks=[checkpoint3]
)

# Plot training & validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(history3.history['accuracy'], label='Train Accuracy')
plt.plot(history3.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model 3 - Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history3.history['loss'], label='Train Loss')
plt.plot(history3.history['val_loss'], label='Validation Loss')
plt.title('Model 3 - Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()


# Evaluate
y_pred3 = model3.predict(X_test)
y_pred_classes3 = np.argmax(y_pred3, axis=-1)
y_true3 = np.argmax(y_test, axis=-1)

accuracy3 = accuracy_score(y_true3, y_pred_classes3)
f1_3 = f1_score(y_true3, y_pred_classes3, average='weighted')

print(f"Model 3 - Accuracy: {accuracy3}")
print(f"Model 3 - F1 Score: {f1_3}")
print(classification_report(y_true3, y_pred_classes3))
cm3 = confusion_matrix(y_true3, y_pred_classes3)

plt.figure(figsize=(8, 6))
sns.heatmap(cm3, annot=True, fmt="d", cmap="Blues", xticklabels=['Damaged', 'No Damage', 'Mild Damage'],
            yticklabels=['Damaged', 'No Damage', 'Mild Damage'])
plt.title('Model 3 - Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

### Training and Evaluation of SegNet Model with Global Average Pooling for Damage Classification

In below code, a custom SegNet-based convolutional neural network is implemented using TensorFlow and Keras to classify structural images into three categories: Damaged, No Damage, and Mild Damage. It has an encoder similar to the SegNet model, and to the decoder but has modified by tying it to a classification task at the end with a Global Average Pooling (GAP). This layer pooling serves to dimensionality reduction and avoids the shape mismatch of the feature maps with the fully connected classifier layer.

The encoder starts with a series of Conv2D and MaxPooling2D with the purpose of reducing spatial dimensions with feature extraction. These features are then passed through Conv2DTranspose layers to give them back more of its spatial structure. Finally, another layer called GlobalAveragePooling2D reduces the dimension of the feature maps to 1D and then class probabilities are computed with the aid of a softmax activated dense layer.

To this end, the model is developed with the Adam optimizer and categorical crossentropy as the loss function effective for multi-class classification tasks. The training is done to the 20th epoch with the batch size of 16 by using the training and the validation data.

Once fitted, the performance of the model is assessed through the test set. The accuracy, F1-Score, classification report, and confusion tables / matrices are then calculated for comparing the predicted values with the true labels. Somewhat related, visualization includes Line plot of training and validation accuracy and loss; Bar chart for the evaluation; Heat map of confusion matrix; where all details of the model’s performance are displayed.

This implementation shows how SegNet, which in its original design is designed for segmentation, can be effectively used for classification by adding GAP layer and be trained and tested for damage assessment using images.










```python
# Define the SegNet model with Global Average Pooling
def segnet(input_size=(256, 256, 3), num_classes=3):
    inputs = tf.keras.Input(input_size)

    # Encoder
    x = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)
    x = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)
    x = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = tf.keras.layers.Conv2DTranspose(256, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)
    x = tf.keras.layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)
    x = tf.keras.layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), activation='relu', padding='same')(x)

    # Global Average Pooling to avoid shape conflicts
    gap = tf.keras.layers.GlobalAveragePooling2D()(x)

    # Classifier
    outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(gap)
    return tf.keras.models.Model(inputs, outputs)

# Compile SegNet
model = segnet()
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=20
)

# Evaluate on the test set
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=-1)  # Predictions
y_true = np.argmax(y_test, axis=-1)          # True labels

# Metrics
accuracy = accuracy_score(y_true, y_pred_classes)
f1 = f1_score(y_true, y_pred_classes, average='weighted')

print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print("Classification Report:")
print(classification_report(y_true, y_pred_classes))
print("Confusion Matrix:")
cm = confusion_matrix(y_true, y_pred_classes)
print(cm)

# Plot Training and Validation Accuracy and Loss
plt.figure(figsize=(12, 6))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Visualizing Evaluation Metrics: Accuracy and F1-Score
metrics = {'Accuracy': accuracy, 'F1 Score': f1}

plt.figure(figsize=(6, 4))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'green'])
plt.title('Evaluation Metrics')
plt.ylabel('Scores')
plt.show()

# Confusion Matrix Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Damaged', 'No Damage', 'Mild Damage'],
            yticklabels=['Damaged', 'No Damage', 'Mild Damage'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

### U-Net Model with Global Average Pooling for Damage Classification
In this implementation, there is the usage of modified U-Net for multi-class classification of structural damages. Situ originally used for image segmentation, the weakness of the U-Net is the collapsibility of feature maps and the ability to maintain hierarchical information of the image due to its encoder-decoder structure and skip connections. However, instead of producing segmentation maps, this U-Net model uses Global Average Pooling (GAP) followed by a dense softmax layer to perform image-level classification into three categories: Damaged, No Damage, and Mild Damage.

The encoder part of the model extracts the features of the input image by applying convolution and max-pooling layers at multiple scales. During the training process, to increase the efficiency, there is a ModelCheckpoint callback implemented which allows saving the model weights that include the highest validation accuracy. GAP shortens the spatial dimensions down to one vector per feature map, making a summary of them before passing the feature through the final classification layer.

During the training phase, a ModelCheckpoint callback is employed in order to save the weights that give the best validation accuracy. The model is trained by Adam optimizer and the appropriate loss function is categorical crossentropy loss as it is used for multi-class classification. It is trained using stochastic gradient descent for 20 epochs on a regular split into training and a validation set with a batch size of 16.

On the provided test set, the performance evaluation is conducted. The above listed model predictions are then passed through an argmax to extract out the best class and compared to true labels. The feature and macro-average F1 indicate total and balanced evaluation of models' accuracy. Furthermore, a confusion matrix heatmap is created to assess the distribution of models based on the classifier across the three categories.

It is then used to plot accuracy and loss for each of the epochs in order to see if there are signs of overfitting or underfitting. In conclusion, this presentation of the U-Net model demonstrates that deep learning segmental networks can be effectively used for binary image classification problems in the area of damage detection while leveraging their inherent feature learning capacities.

```python
from tensorflow.keras.callbacks import ModelCheckpoint

def unet(input_size=(256, 256, 3)):
    inputs = tf.keras.Input(input_size)
    c1 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    p1 = tf.keras.layers.MaxPooling2D((2, 2))(c1)

    c2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    p2 = tf.keras.layers.MaxPooling2D((2, 2))(c2)

    c3 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
    p3 = tf.keras.layers.MaxPooling2D((2, 2))(c3)

    c4 = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(p3)
    c4 = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(c4)

    u5 = tf.keras.layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c4)
    u5 = tf.keras.layers.concatenate([u5, c3])
    c5 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(u5)
    c5 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c5)

    u6 = tf.keras.layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = tf.keras.layers.concatenate([u6, c2])
    c6 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u6)
    c6 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c6)

    u7 = tf.keras.layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = tf.keras.layers.concatenate([u7, c1])
    c7 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u7)
    c7 = tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c7)

    gap = tf.keras.layers.GlobalAveragePooling2D()(c7)
    outputs = tf.keras.layers.Dense(3, activation='softmax')(gap)

    return tf.keras.models.Model(inputs=[inputs], outputs=[outputs])

model4 = unet()
model4.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Add ModelCheckpoint
checkpoint4 = ModelCheckpoint(
    filepath='best_model4.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

history4 = model4.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=20,
    callbacks=[checkpoint4]
)

# Plot training & validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(history4.history['accuracy'], label='Train Accuracy')
plt.plot(history4.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model 4 - Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history4.history['loss'], label='Train Loss')
plt.plot(history4.history['val_loss'], label='Validation Loss')
plt.title('Model 4 - Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

#evaluation
segmented_masks4 = model4.predict(X_test)
y_pred4 = np.argmax(segmented_masks4, axis=1)
y_true4 = np.argmax(y_test, axis=1)

accuracy4 = accuracy_score(y_true4, y_pred4)
f1_4 = f1_score(y_true4, y_pred4, average='weighted')

print(f"Model 4 - Accuracy: {accuracy4}")
print(f"Model 4 - F1 Score: {f1_4}")
print(classification_report(y_true4, y_pred4))
cm4 = confusion_matrix(y_true4, y_pred4)

plt.figure(figsize=(8, 6))
sns.heatmap(cm4, annot=True, fmt="d", cmap="Blues", xticklabels=['Damaged', 'No Damage', 'Mild Damage'],
            yticklabels=['Damaged', 'No Damage', 'Mild Damage'])
plt.title('Model 4 - Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()


```

```python
def compare_models(metrics_list):
    """
    metrics_list: List of dicts with keys: 'name', 'accuracy', 'f1'
    """
    names = [m['name'] for m in metrics_list]
    accuracies = [m['accuracy'] for m in metrics_list]
    f1_scores = [m['f1'] for m in metrics_list]

    # Bar Chart
    plt.figure(figsize=(10, 5))
    x = np.arange(len(names))
    plt.bar(x - 0.2, accuracies, width=0.4, label='Accuracy', color='blue')
    plt.bar(x + 0.2, f1_scores, width=0.4, label='F1 Score', color='green')
    plt.xticks(x, names)
    plt.ylabel('Scores')
    plt.title('Model Comparison: Accuracy vs F1 Score')
    plt.legend()
    plt.show()

    # Line Plot
    plt.figure(figsize=(10, 5))
    plt.plot(names, accuracies, marker='o', label='Accuracy', color='blue')
    plt.plot(names, f1_scores, marker='s', label='F1 Score', color='green')
    plt.title('Model Comparison')
    plt.ylabel('Scores')
    plt.legend()
    plt.show()

    # Histogram
    plt.figure(figsize=(10, 5))
    plt.hist([accuracies, f1_scores], label=['Accuracy', 'F1 Score'], color=['blue', 'green'], bins=np.linspace(0, 1, 11), alpha=0.7)
    plt.title('Histogram of Accuracies and F1 Scores')
    plt.xlabel('Scores')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Metrics for all models (updated with actual names)
metrics_all = [
    {'name': 'FCN', 'accuracy': accuracy, 'f1': f1},
    {'name': 'DeepLab', 'accuracy': accuracy2, 'f1': f1_2},
    {'name': 'SegNet', 'accuracy': accuracy3, 'f1': f1_3},
    {'name': 'U-Net', 'accuracy': accuracy4, 'f1': f1_4},
]

compare_models(metrics_all)

# Metrics for all models
metrics_all = [
    {'name': 'Model 1', 'accuracy': accuracy, 'f1': f1},            # from Model 1
    {'name': 'Model 2', 'accuracy': accuracy2, 'f1': f1_2},
    {'name': 'Model 3', 'accuracy': accuracy3, 'f1': f1_3},
    {'name': 'Model 4', 'accuracy': accuracy4, 'f1': f1_4},
]

compare_models(metrics_all)
```
